#### 2.1. Programa que retorna $e$ en base a la suma infinita

````python
import numpy as np
n_elem = input("Ingresar número de términos de la aproximación:")
try:
    n_elem = int(n_elem)
    assert n_elem >= 1, "El número ingresado debería ser un entero mayor a 0"
    terms = [1.0/np.math.factorial(i-1) for i in range(1,n_elem+1)]
    approx = np.sum(terms)
    print('e es aproximadamente:',approx)
except:
    print("El número ingresado debería ser un entero mayor a 0")
````

#### 3.1. Generar matriz aleatoria y calcular métrica

````python
n = input("Ingresar orden de la matriz cuadrada:")
try:
    n = int(n)
    assert n >= 1, "El número ingresado debería ser un entero mayor a 0"
    C = np.random.rand()*np.random.randn(n,n)+np.random.rand()
    trace = np.eye(n) * C # Obtenemos elementos de la diagonal principal
    trace = np.sum(trace) # Todos los elementos fuera de la diagonal principal se volvieron 0s
    denominator = np.sum(C) # La doble sumatoria corresponde a acumular todos los elementos de la matriz
    metric = trace/denominator
    print('matriz:\n',C)
    print('traza:',trace)
    print('denominador',denominator)
    print('metrica:',metric)
except:
    print("El número ingresado debería ser un entero mayor a 0")
````

#### 3.2. Generar señal aleatoria a partir de la edad y graficar la señal

````python
import numpy as np
edad = 24
senal=np.empty((0,))
matriz=np.empty((0,100))
for i in range(edad):
    ventana=np.random.rand()*np.random.randn(100)+np.random.rand()
    senal=np.hstack((senal,ventana)) # Para concatenar en un vector
    matriz=np.vstack((matriz,ventana)) # Para concatenar en una matriz
    
import matplotlib.pyplot as plt # libreria usada para graficas
%matplotlib inline

def plot_signal(x,y=None,xlabel=None,ylabel=None,size=None):
    fig, ax = plt.subplots(1, 1)  
    ax.set_aspect('auto')

    if y is not None:
        ax.plot(x,y)
    else:
        ax.plot(x)
    if xlabel is not None:
        plt.xlabel(xlabel,fontsize=18)
    if ylabel is not None:
        plt.ylabel(ylabel,fontsize=18)
    if size is not None:
        fig.set_size_inches(size)
    plt.grid()
    plt.tight_layout()
    plt.show()

plot_signal(senal,xlabel='muestras',ylabel='valores',size=(20,5))

for w in range(matriz.shape[0]):
    dev_std = np.std(matriz[w,:])
    mean = np.mean(matriz[w,:])
    print('\nventana:',w+1)
    print('media:',mean)
    print('dev_std:',dev_std)
````

##### ¿Como se relacionan estos valores con la grafica que generó anteriormente?
En esencia son una aproximación de las medias y desviaciones estándar que generaron cada una de las de las señales aleatorias de cada ventana. Esto es debido a la siguiente linea de código que generó las señales:

```python
ventana=np.random.rand()*np.random.randn(100)+np.random.rand()
```

Estas lineas corresponden a una distribución normal gaussiana de acuerdo a la siguiente relación obtenida de la [documentación de numpy](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randn.html):

$\mathcal{N}(\mu,\,\sigma^{2})=$ `sigma * np.random.randn(...) + mu`

Ya desde una vista más desde las señales en sí; la media la podemos pensar como el valor DC de cada ventana de 100 puntos y la desviación estándar la amplitud con la que la señal se desvía de esa media. Así se tiene que:

- Mientras mayor la media, más arriba en el eje vertical quedara centrada la ventana correspondiente.
- Mientras mayor la desviación estándar más "gorda" se verá en la gráfica.


#### 3.3. Implementación en funciones del numeral anterior

````python
import numpy as np
import matplotlib.pyplot as plt # libreria usada para graficas
%matplotlib inline

edad = 24

def make_signal(num_windows,num_frames=100):
    senal=np.empty((0,))
    matriz=np.empty((0,num_frames))
    for i in range(num_windows):
        ventana=np.random.rand()*np.random.randn(num_frames)+np.random.rand()
        senal=np.hstack((senal,ventana)) # Para concatenar en un vector
        matriz=np.vstack((matriz,ventana)) # Para concatenar en una matriz
    return senal,matriz


def plot_signal(x,y=None,xlabel=None,ylabel=None,size=None):
    fig, ax = plt.subplots(1, 1)  
    ax.set_aspect('auto')

    if y is not None:
        ax.plot(x,y)
    else:
        ax.plot(x)
    if xlabel is not None:
        plt.xlabel(xlabel,fontsize=18)
    if ylabel is not None:
        plt.ylabel(ylabel,fontsize=18)
    if size is not None:
        fig.set_size_inches(size)
    plt.grid()
    plt.tight_layout()
    plt.show()


def get_stats(senal,matriz):
    plot_signal(senal,xlabel='muestras',ylabel='valores',size=(20,5))    
    dev_std = np.std(matriz,axis=-1) # vector columna
    dev_std = np.expand_dims(dev_std, axis=-1)
    mean = np.mean(matriz,axis=-1) # vector columna
    mean = np.expand_dims(mean, axis=-1)
    vector = np.hstack((mean,dev_std)) # columna 0: medias, columna 1: desviaciones estandar
    return vector

senal,matriz = make_signal(edad)
vector = get_stats(senal,matriz)
print('vector [medias,desviaciones]\n',vector)
````
